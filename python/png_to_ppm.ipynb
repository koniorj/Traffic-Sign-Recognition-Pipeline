```python
import pandas as pd
from PIL import Image
import os

csv_file = r"data_gtsrb\Test.csv"
source_folder = r"data_gtsrb"
dest_folder = r"ppm_test"
limit = 1000

os.makedirs(dest_folder, exist_ok=True)

df = pd.read_csv(csv_file)
subset = df.head(limit)
truth_data = []

for index, row in subset.iterrows():
    img_path = os.path.join(source_folder, row['Path'])
    base_name = os.path.basename(row['Path']).replace('.png', '.ppm')
    img = Image.open(img_path)
    img.save(os.path.join(dest_folder, base_name), "PPM")
    truth_data.append({'filename': base_name, 'class_id': row['ClassId']})

pd.DataFrame(truth_data).to_csv(os.path.join(dest_folder, "ground_truth.csv"), index=False)
```
